```python
import pandas as pd
from masakali_data.utils import open_csv, save_csv, current_directory

current_directory = current_directory()
```

```python
website = open_csv(current_directory, 'website/2022/website_revenue_idr_2022.csv')
website = website[['check_in', 'check_out', 'nights', 'villa_name', 'platform']]
website.head()
```

```python
airbnb = open_csv(current_directory, 'airbnb/2022/airbnb_revenue_idr_2022.csv')
def rename_villas(df):
    # Define the mapping dictionary
    name_map = {
        'Chandra: Entire villa w/ private pool and kitchen': 'chandra',
        'Surya: Luxury Villa with kitchen and private pool': 'surya',
        'Lakshmi: Luxury 2 bedroom villa': 'lakshmi',
        'Jala: Entire villa with private pool and kitchen': 'jala',
        'Akasha: 3 bedroom luxury villa': 'akasha'
    }

    # Use the map function to replace the names
    df['villa_name'] = df['villa_name'].map(name_map)

    return df
airbnb = rename_villas(airbnb)

airbnb = airbnb[['check_in', 'check_out', 'nights', 'villa_name', 'platform']]
airbnb.head()
```

```python
agoda = open_csv(current_directory, 'agoda/2022/agoda_revenue_idr_2022.csv')
agoda = agoda[['check_in', 'check_out', 'nights', 'villa_name', 'platform']]
agoda.head()
```

```python

booking = open_csv(current_directory, 'booking/2022/booking_revenue_idr_2022.csv')


booking = booking[['check_in', 'check_out', 'nights', 'villa_name', 'platform']]
booking.head()
```

```python
combined_cols = [
    'check_in',
    'check_out',
    'nights',
    'villa_name',
    'platform'
]

df = pd.concat([website, airbnb, agoda, booking], ignore_index=True)
df.to_csv('combined_bookings_2022.csv', index=False)
```
