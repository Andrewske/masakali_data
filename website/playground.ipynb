```python


def parse_website_data(year):
    df = pd.read_csv('website_and_direct_accurate_price.csv')

    df = df.astype({
         "check_in": "datetime64[ns]",
         "check_out": "datetime64[ns]",
    })
    df = df[df.check_out.dt.year == year]

    print(len(df))


    df = df.rename(columns={
        'id': 'booking_id',
        'price': 'gross_income',
        'commission_included': 'commission',
        'channel_name': 'platform',
    })

    df['net_income'] = df.gross_income - df.commission
    df['taxable_amount'] = df.net_income

    df_idr_small = df[df.currency == 'idr'].copy()
    df_usd_small = df[df.currency == 'usd'].copy()


    df_idr = pd.concat([df_idr_small, convert_columns_to_idr(df_usd_small, ['gross_income', 'commission', 'net_income', 'taxable_amount'])]).reset_index()
    df_usd = pd.concat([df_usd_small, convert_columns_to_usd(df_idr_small, ['gross_income', 'commission', 'net_income', 'taxable_amount'])]).reset_index()

    columns = ['booking_id', 'date', 'guest_name', 'villa_name', 'gross_income'] #'commission', 'net_income', 'taxable_amount']

    df_idr = df_idr[columns]
    df_usd = df_usd[columns]

    

    


parse_website_data(2022)

```
